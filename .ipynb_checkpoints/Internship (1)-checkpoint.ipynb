# Random Forest

```python
# Use numpy to convert to arrays
import numpy as np
# Labels are the values we want to predict
inactive = np.array(air['IS_INACTIVE'])
# Remove the labels from the features
# axis 1 refers to the columns
#air= air.drop('actual', axis = 1)
feature_list = list(air.columns)
# Convert to numpy array
air = np.array(air)
```

```python
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
train_air, test_air, train_Inactive, test_Inactive = train_test_split(air,inactive, test_size = 0.25, random_state = 42)
print('Training data Shape:', train_air.shape)
print('Training inactive Shape:', train_Inactive.shape)
print('Testing data Shape:', test_air.shape)
print('Testing inactive Shape:', test_Inactive.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(train_air, train_Inactive);
```

```python
predictions = rf.predict(test_air)
errors = abs(predictions - test_Inactive)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
mape = 100 * (errors / test_Inactive)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```
